```python
def swap(inp):
    l = 0
    r =len(inp)-1
    while l<r:
        inp[l], inp[r] = inp[r], inp[l]
        l+=1
        r-=1
    return inp
```

```python
swap("ABC123")
```

```python
# Add a demonstration using the example input
inputs = ["ABC123", "XYZ789", "PQR555"]
outputs = [swap(s) for s in inputs]
print(outputs)  # Expected: ['321CBA', '987ZYX', '555RQP']
```
